## 3) 그림 범위 지정

그림의 범위를 수동으로 지정하려면 xlim 명령과 ylim 명령을 사용한다. 이 명령들은 그림의 범위가 되는 x축, y축의 최소값과 최대값을 지정한다.

````python
plt.title('set limit')
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25],
         c = 'm', lw ='5', ls=':', marker='1', ms='10', mec ='g',
         mew='5', mfc='r')
plt.xlim(0, 10)
plt.ylim(-5, 30)
plt.show()
````

## 4) 틱 설정
플롯이나 차트에서 X축, Y축상의 위치 표시 지점을 틱(tick)이라 하고 여기에 쓰여진 숫자 혹은 글자를 틱 라벨(tick label)이라 한다. 보통 자동으로 정해지지만, 수동으로 설정하고 싶다면 xticks이나 yticks을 사용한다.

````python
import numpy as np
````

````python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title('set tick label')
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi/2, 0 , np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
````

틱 라벨 문자열에는 $$ 사이에 LaTeX 수학 문자식을 넣을 수도 있다.


````python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("set tick label by doing LaTeX")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
````

## 5) 그리드 설정

````python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("set tick label by doing LaTeX")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])

plt.grid(True)
plt.show()
````

## 6) 여러개의 선 그리기, 겹쳐 그리기

한 줄로 여러 개의 선을 그릴 수 있으나, 헷갈리니까 복수의 줄을 사용해 겹쳐 그리는 것이 낫다.

````python
plt.title('lots of plot')
plt.plot([1, 4, 9, 16],
         c='b', lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.plot([1, 2, 4, 8, 16, 25],
         c='k', lw=3, ls=':', marker='s', ms=10, mec='m', mew=5, mfc='c')
plt.grid(True)
plt.show()
````

## 7) 범례(legend)

여러개의 라인 플롯을 동시에 그리는 경우에 각 선이 무슨 자료인지를 표시하는 legend(범례)를 추가할 수 있다. 범례의 위치는 자동으로 정해지지만 수동으로 설정하고 싶으면 loc를 사용한다.

````python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("using legend")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc='best')
plt.show()
````

## 8) X축, Y축 라벨 타이틀

플롯의 x축 위치와 y축 위치에는 각각 그 데이터가 의미하는 바를 표시하기 위해 라벨(label)를 추가할 수 있다. 라벨을 붙이려면 xlabel. ylabel 명령을 사용한다. 또 플롯의 위에는 title 명령으로 제목(title)을 붙일 수 있다.





````python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.show()
````

````python
X = range(10)
Y = np.linspace(1, 100, 10)
Z = Y ** 1.1
plt.plot(X, Y, c='r', marker='^', ls='--', label='Y')
plt.plot(X, Z, c='b', marker='o', ls=':', label='Z')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(loc='best')
plt.show()
````
